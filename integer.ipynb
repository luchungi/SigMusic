```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
model = torch.nn.Sequential(
    torch.nn.Linear(1, 64),
    torch.nn.ReLU(),
    torch.nn.Linear(64, 64),
    torch.nn.ReLU(),
    torch.nn.Linear(64, 1)
)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
def f(x):
    return x - torch.sin(2*torch.pi*x) / (2*torch.pi)
```

```python
f(torch.tensor(4.5))
```

```python
for i in range(1000):
    y = torch.randint(0, 5, (64, 1)).float()
    x = y + torch.randn(64, 1)*0.2
    # x = torch.randn(64, 1)
    pred = model(x)
    pred = f(pred)
    # loss = torch.nn.MSELoss()(torch.round(y.detach()), y) + torch.nn.MSELoss()(torch.randint(0, 5, (64,1)).float(), y)
    loss = torch.nn.MSELoss()(pred, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    # plt.plot(np.arange(64), y.detach().numpy())
    # plt.show()
    print(loss.mean().item())
```

```python
plt.plot(np.arange(64), pred.detach().squeeze())
plt.plot(np.arange(64), y.detach().squeeze())
```
